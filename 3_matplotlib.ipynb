## How to use matplotlib
matplotlibはさまざまな可視化ツールを提供します．

##インターフェイス
matplotlibでは二つのインターフェイスが存在しています．


*   MATLABスタイル
*   オブジェクト指向スタイル



```python
%matplotlib inline
import matplotlib as mlp
import matplotlib.pyplot as plt
plt.style.use('classic')
import numpy as np
```

```python
fig = plt.figure()  # 軸，グラフィックス，テキスト，ラベルなどを含むキャンバス
ax = plt.axes()   # メモリとラベルをもつ境界
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

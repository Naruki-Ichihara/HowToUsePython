# How to use pandas

## はじめに
データサイエンスにおける主たるデータ処理を効率的に行うインターフェイスを提供します．

## Seriesオブジェクト
Seriesオブジェクトはvalues属性とindex属性を有するpandasの基本的なオブジェクトです．
values属性は1次元numpy配列であり， index属性により拡張されたラベル付き1次元numpy配列と捉えることができます．

```python
import pandas as pd
series = pd.Series([0, 1, 2, 3, 4, 5])
series.values
```

```python
series.index
```

## DataFrameオブジェクト
RowIndexとColumnIndexを有するマトリクス形式のデータ構造で，pandasの基本的なオブジェクトです．
複数のSeriesをColumu方向に（あるいはrow方向に）スタックすることで生成できます．あるいは複数のSeriesをIndex付きで保持した特殊な辞書構造です．

```python
series_0 = pd.Series([0, 1, 2, 3, 4])
series_1 = pd.Series([5, 6, 7, 8, 9])
data = pd.DataFrame([series_0, series_1], index=[0, 1])
data
```

## Indexオブジェクト
Indexオブジェクトは不変かつ順序付きの多重集合体です．多くの一般的な配列演算が定義されていますが，一般的な方法で要素を更新することができません．

```python
index = pd.Index([2, 3, 3, 4, 5, 7])
index
```

```python
index[2] = 0
```

indexオブジェクトは種々のブーリアン演算が定義されています．

```python
index_0 = pd.Index([0, 1, 3, 5, 7, 9])
index_1 = pd.Index([2, 3, 5, 7, 11])
```

```python
# Intersection
index_0 & index_1
```

```python
# Union
index_0 | index_1
```

```python
# symmetric diffrence
index_0 ^ index_1
```
